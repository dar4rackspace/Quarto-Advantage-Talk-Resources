## What is Quarto?

![](images/confused-white-persian-guardian.gif){fig-align="center"}

## TLDR {auto-animate=true}

- Quarto enables you to weave together content and executable code into a finished document.

- Highlight Quarto's seamless integration with Python, R, Julia, and Observable, enabling dynamic content generation and interactive data exploration

![rstudio-qmd-how-it-works.png (1400Ã—257) (quarto.org)](https://quarto.org/docs/get-started/hello/images/rstudio-qmd-how-it-works.png)

## Running Code

When you click the **Render** button a document will be generated that includes both content and the output of embedded code. You can embed code like this:


````python
#| echo: true
1 + 1
````

::: aside
[Check other execution options](https://quarto.org/docs/computations/execution-options.html)

```{mermaid}
  E(pandoc) --> F(HTML)
```

```{.yml code-line-numbers="4-9"}
  navbar:
    logo: "images/Rackspace_Technology_Logo_RGB_RED.png"
```

### Sidebar Set Up

